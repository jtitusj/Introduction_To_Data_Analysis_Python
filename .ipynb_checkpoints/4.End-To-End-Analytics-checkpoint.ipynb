#### Distribution Analysis

Now that we are familiar with basic data characteristics, let us study distribution of various variables.

```python
# 8. Plot histograms for the different variables. 
# dims variable is already set below to improve clarity of resulting plots.
# If necessary, create individual histograms for each variable.
# try to change the bins parameter in .hist to see how it affects the visualization
dims = (14, 6)
df_train.hist(figsize=dims, bins=50)
```

What do the histograms tell us?

```python
# 9. Create boxplot for ApplicantIncome
df_train.boxplot('ApplicantIncome', return_type='axes')
```

This confirms the presence of a lot of outliers/extreme values. This can be attributed to the income disparity in the society. Part of this can be driven by the fact that we are looking at people with different education levels. Let us segregate them by Education:

```python
# 10. create boxpot for ApplicantIncome but this time, group by education.
# This can be done by using the 'by' parameter in .boxplot()

df_train.boxplot('ApplicantIncome', by='Education', return_type='axes')
```

```python
# 11. Is there a difference in income between the two genders?
df_train.boxplot('ApplicantIncome', by='Gender', return_type='axes')
```

We can see that there is no substantial different between the mean income of males and females. But there are a higher number of males with very high incomes, which are appearing to be the outliers.
